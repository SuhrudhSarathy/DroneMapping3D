```python
import numpy as np
import matplotlib.pyplot as plt
from functools import partial
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

# skimage
import skimage
from skimage.transform import rotate
from skimage.util import random_noise
from skimage.filters import gaussian
import skimage.io as io

import os
tqdm = partial(tqdm, position=0, leave=True)

main_directory = "/content/drive/MyDrive/Aero_Forest_Project/Datasets"
augmented_directory = "/content/drive/MyDrive/Aero_Forest_Project/Augmented_Datasets"
```

```python
# List the directory
classes = os.listdir(main_directory)
classes
```

```python
# make the directories for the specific classes
for class_ in classes:
    try:
        os.mkdir(f"{augmented_directory}/{class_}")
    except Exception as e:
        print("Folders Already Exist, Delete them")
        break
```

```python
def augement_images(image_path):
    outname = image_path.replace("Datasets", "Augmented_Datasets")
    image = io.imread(image_path)
    # image = image/255 # normalise the image

    # Rotate image
    img1 = rotate(image, 45, mode="wrap")
    img2 = rotate(image, -45, mode="wrap")

    # Add random noise
    img_rnd = random_noise(image, var=1**2)
    # Gaussian blur
    img_gaussian = gaussian(image, sigma=5, multichannel=True)

    # Flip the image
    img_left_flip = np.fliplr(image)
    
    img_up_down_flip = np.flipud(image)

    io.imsave(f"{outname}img.png", image)
    io.imsave(f"{outname}img2.png", img1)
    io.imsave(f"{outname}img1.png", img1)
    io.imsave(f"{outname}img_rnd.png", img1)
    io.imsave(f"{outname}img_gaussian.png", img1)
    io.imsave(f"{outname}img_left_flip.png", img1)
    io.imsave(f"{outname}img_up_down_flip.png", img1)

    # plt.imshow(img_rnd)
```

```python
augement_images("/content/drive/MyDrive/Aero_Forest_Project/Datasets/Acanthus Ilicifolius/AI-1.jpg")
```

```python
# Traverse through all the images and get this done
from google.colab import output
for cat in tqdm(classes):
    imgs = os.listdir(f"{main_directory}/{cat}")
    for img in imgs:
        path = f"{main_directory}/{cat}/{img}"
        augement_images(path)
```
